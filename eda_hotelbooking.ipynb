```python
guest_country = data[data['is_canceled']==0] ['country'].value_counts().reset_index()
guest_country.columns=['country', 'numberofguests']
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.express as px 
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
total_guests = guest_country['numberofguests'].sum()
guest_country['guest in %'] = round(guest_country['numberofguests']/total_guests*100,2)
```

number of guests in percentage 
    

```python
trace = go.Bar(
    x=guest_country['country'],
    y=guest_country['guest in %'],
    marker=dict(color='#C0911B')
)
data1 = [trace]
layout = go.Layout(
    title='guest in percentage'
)
fig = go.Figure(data=data1,layout=layout)
pyo.plot(fig)
```

```python
prices = valid_guest[['hotel','reserved_room_type','adr_updated']].sort_values('reserved_room_type')

plt.figure(figsize=(12,8))
sns.boxplot(x='reserved_room_type',
            y='adr_updated',
            hue='hotel',
            data=prices)

plt.title('prices of the room per person per night', fontsize=16)
plt.xlabel('room type', fontsize=16)
plt.ylabel('price[EUR]', fontsize=16)

plt.ylim(0,160)
plt.show()
```
